```python
import pandas as pd
encoding_format = "latin1"
channels = pd.read_csv("/Users/juraev_samoyiddin/JupyterProjects/archive/channels.csv", encoding=encoding_format)
deliveries = pd.read_csv("/Users/juraev_samoyiddin/JupyterProjects/archive/deliveries.csv", encoding=encoding_format)
drivers = pd.read_csv("/Users/juraev_samoyiddin/JupyterProjects/archive/drivers.csv", encoding=encoding_format)
hubs = pd.read_csv("/Users/juraev_samoyiddin/JupyterProjects/archive/hubs.csv", encoding=encoding_format)
orders = pd.read_csv("/Users/juraev_samoyiddin/JupyterProjects/archive/orders.csv", encoding=encoding_format)
payments = pd.read_csv("/Users/juraev_samoyiddin/JupyterProjects/archive/payments.csv", encoding=encoding_format)
stores = pd.read_csv("/Users/juraev_samoyiddin/JupyterProjects/archive/stores.csv", encoding=encoding_format)

df = orders.merge(stores, on="store_id", how="inner")
df = df.merge(channels, on="channel_id", how="inner")
df = df.merge(payments, on="payment_order_id", how="inner")
df = df.merge(deliveries, on="delivery_order_id", how="inner")
df = df.merge(drivers, on="driver_id", how="inner")
df = df.merge(hubs, on="hub_id", how="inner")

threshold_columns = int(0.3 * len(df)) 
df = df.dropna(axis=1, thresh=threshold_columns)

threshold_rows = int(0.3 * len(df.columns))
df = df.dropna(thresh=threshold_rows)

df.fillna(df.median(numeric_only=True), inplace=True)

from IPython.display import display
display(df)
```
